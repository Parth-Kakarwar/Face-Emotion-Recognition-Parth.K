```python
import cv2
import matplotlib.pyplot as plt
```

```python
from deepface import DeepFace
```

```python
img = cv2.imread('happyboy.jpg')
```

```python
plt.imshow(img)  ##By default BGR
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
predictions=DeepFace.analyze(img)
```

```python
predictions
```

```python
type(predictions)
```

```python
predictions[0]['dominant_emotion']
```

```python
# Loading the Haar Cascade for face detection
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Checking if the cascade file loaded correctly
if faceCascade.empty():
    print("Error loading cascade file")
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # For Detecting faces in the grayscale image
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Drawing a rectangle around the faces
    for x, y, w, h in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
# Using put Text() method for
# inserting text on video
cv2. putText (img,
predictions[0] ['dominant_emotion'] ,
(50, 50),
font, 1,
(0, 0, 255),
2,
cv2.LINE_4);
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
